```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/drive/MyDrive/car details v4.csv")
data.head()
```

```python
numerical_cols = ['Length', 'Width', 'Height', 'Seating Capacity', 'Fuel Tank Capacity']
for col in numerical_cols:
    data[col].fillna(data[col].median(), inplace=True)
```

```python
categorical_cols = ['Engine', 'Max Power', 'Max Torque', 'Drivetrain']
for col in categorical_cols:
    data[col].fillna('Unknown', inplace=True)
```

```python
data = data[data['Kilometer'] <= 500000]
```

```python
import re

def extract_number(value):
    num = re.findall(r'[\d.]+', str(value))
    return float(num[0]) if num else None

data['Engine'] = data['Engine'].apply(extract_number)
data['Max Power'] = data['Max Power'].apply(extract_number)
data['Max Torque'] = data['Max Torque'].apply(extract_number)
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='Price', data=data)
plt.title('Distribusi Harga Mobil')
plt.show()
```

```python
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
```

```python
import re

def extract_number(value):
    num = re.findall(r'[\d.]+', str(value))
    return float(num[0]) if num else None

data['Engine'] = data['Engine'].apply(extract_number)
data['Max Power'] = data['Max Power'].apply(extract_number)
data['Max Torque'] = data['Max Torque'].apply(extract_number)
```

```python
# 4. Visualisasi Data
# Visualisasi Distribusi Harga Kendaraan
plt.figure(figsize=(8, 5))
sns.histplot(data['Price'], bins=50, kde=True)
plt.title('Distribusi Harga Kendaraan')
plt.xlabel('Harga (INR)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Visualisasi Hubungan Tahun Produksi dan Harga Kendaraan
plt.figure(figsize=(12, 6))
sns.boxplot(x='Year', y='Price', data=data)
plt.xticks(rotation=45)
plt.title('Pengaruh Tahun Produksi terhadap Harga Kendaraan')
plt.show()
```

```python
# Visualisasi Hubungan Kilometer Tempuh dan Harga Kendaraan
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Kilometer', y='Price', data=data)
plt.title('Hubungan Kilometer Tempuh terhadap Harga Kendaraan')
plt.xlabel('Kilometer')
plt.ylabel('Harga (INR)')
plt.show()
```
